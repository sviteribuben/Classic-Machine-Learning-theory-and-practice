# Классическое машинное обучение (Classical Machine Learning)
Используется когда есть полный доступ к набору данных и имеется явный набор признаков (фич)

* **ОБУЧЕНИЕ С УЧИТЕЛЕМ (with supervising)**
существует большая выборка данных, на основании которой нужно чтото предсказать, и набор гипотез. Данные уже отсортированы по категориям или имеются конкретные значения величин. Есть данные и есть идеальный исход для них. Такие алгоритмы решают две основные задачи - **регрессии и классификации**

* **ОБУЧЕНИЕ БЕЗ УЧИТЕЛЯ** 
нет размеченных данных для обучения. Целью является нахождение закономерностей в данных или их аномалий и их группировка. Для данных нет идеального выхода

## Задача регрессии

## Обучение с подкреплением

Применяется в случае когда мы не имеем доступа к данным, но можем взаимодействовать с окружающей средой. 

Основная идея - выживание в среде. Цель - Минимизация ошибки а не предсказание всех ходов.

**Обучение с подкреплением** использует систему **поощрений и штрафов**

# Ансамбли алгоритмов

Одним из способов повышения точности моделей - является создание и обучение ансамблей моделей. - Наборов моделей созданных для решения одной и той же задачи с последующим объединение их результатов прогнозирования в **единый прогноз** 

Для создания ансамблей можно использовать любые алгоритмы но мы получим лучшие результаты если будем использовать нестабильные алгоритмы

##СТЭКИНГ (Staking)
каждый набор данных мы используем как входной параметр разных алгоритмов. При этом алгоритмы находятся в равных условиях так как получают один и тот же набор данных.

## БЭГИНГ (Baging)

Его суть в том что алгоритм испольлзуем один и тот же, а наборы данных меняем. ДАнные компонуются случайным образом.

## БУСТИНГ (BOOSTING)
Используется один и тот жэ класс алгоритмов. Исправляем ошибки на каждой итерации. Следующий набор данных содержит больше экземпляров-ошибок. Роль корректора. Чем больше итераций - тем лучше будут исправлятся ошибки совершенные предыдущими ансамблями.
